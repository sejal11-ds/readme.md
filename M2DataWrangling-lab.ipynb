## Imputing missing values


Find the  value counts for the column WorkLoc.


```python
# your code goes here
import pandas as pd

# Load the data from the CSV file
url = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBM-DA0321EN-SkillsNetwork/LargeData/m1_survey_data.csv"
df = pd.read_csv(url)

# Get the value counts for the 'WorkLoc' column
workloc_counts = df['WorkLoc'].value_counts()

print(workloc_counts)
```

Identify the value that is most frequent (majority) in the WorkLoc column.


```python
#make a note of the majority value here, for future reference
import pandas as pd

# Load the data from the CSV file
url = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBM-DA0321EN-SkillsNetwork/LargeData/m1_survey_data.csv"
df = pd.read_csv(url)

# Get the value counts for the 'WorkLoc' column
workloc_counts = df['WorkLoc'].value_counts()

# Identify the most frequent value
most_frequent_workloc = workloc_counts.idxmax()

print(f"The most frequent value in the 'WorkLoc' column is: {most_frequent_workloc}")
```

Impute (replace) all the empty rows in the column WorkLoc with the value that you have identified as majority.


```python
# your code goes here
import pandas as pd

# Load the data from the CSV file
url = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBM-DA0321EN-SkillsNetwork/LargeData/m1_survey_data.csv"
df = pd.read_csv(url)

# Identify the most frequent value in the 'WorkLoc' column
most_frequent_workloc = df['WorkLoc'].mode()[0]

print(f"The most frequent value in the 'WorkLoc' column is: {most_frequent_workloc}")

# Replace missing values in the 'WorkLoc' column with the most frequent value
df['WorkLoc'].fillna(most_frequent_workloc, inplace=True)

# Verify that there are no missing values in the 'WorkLoc' column
missing_workloc = df['WorkLoc'].isnull().sum()
print(f"Number of missing values in 'WorkLoc' column after imputation: {missing_workloc}")

```
